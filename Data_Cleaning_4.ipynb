# **Character Encoding**

```python
import pandas as pd
import numpy as np

import charset_normalizer
pd.read_csv('/content/sample_data/california_housing_train.csv')
np.random.seed(0)
```

```python
before = 'This is a symbol of Euro: €'
type(before)
```

```python
after = before.encode('utf_8', errors='replace')
type(after)
```

```python
after
```

```python
print(after.decode('utf-8'))
```

```python
# if we encode 'before' in ascii using errors = 'replace',
# it will replace euro sign with a placeholder
# but when i decode it again in utf_8 the euro sign would have been lost coz there was no euro sign in ascii.
after = before.encode('ascii', errors='replace')

after.decode('utf_8', errors='replace')
```

```python
# detecting file encoding
with open('/content/sample_data/california_housing_train.csv', 'rb') as rawdata:
  result = charset_normalizer.detect(rawdata.read(10000))
  print(result)
```
