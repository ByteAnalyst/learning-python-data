```python
# importing pandas and numpy libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
# loading csv file
district_prices = pd.read_csv('/content/District-prices-monthly.csv')
district_prices.head()
```

```python
# displaying column names
district_prices.columns.tolist()
```

```python
# standarizing column names by replacing - and capatalizing all letters
district_prices.columns = district_prices.columns.str.replace('_', ' ')
district_prices.columns = district_prices.columns.str.upper()
```

```python
# displaying columns names after applying changes
district_prices.columns.tolist()
```

```python
# checking for duplicated values
district_prices.duplicated().sum()
```

```python
# checking for information about columns. which show us datatypes of columns and number of non null values
district_prices.info()
```

```python
# we will check missing values
district_prices.isnull().sum()
```

As you can see only 2 columns have missing values. We will deal with them when cleaning that particular column. Now we will start with 1st column and then move to next one by one.

```python
# changing the datatype of 1st column to datetime as the column is representing date
district_prices['YEAR MONTH'] = pd.to_datetime(district_prices['YEAR MONTH'], errors='coerce')
```

```python
print(district_prices['YEAR MONTH'].dtype)
```

```python
# changing the name of column from YEAR MONTH TO DATE
district_prices.rename(columns= {'YEAR MONTH': 'DATE'}, inplace= True)
```
