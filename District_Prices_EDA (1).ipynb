```python
# importing pandas and numpy libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
# loading csv file
district_prices = pd.read_csv('/content/District-Prices-Monthly.csv')
district_prices.head()
```

```python
# standarizing column names by replacing - and capatalizing all letters
district_prices.columns = district_prices.columns.str.replace('_', ' ')
district_prices.columns = district_prices.columns.str.upper()
```

```python
# changing the datatype of 1st column to datetime as the column is representing date
district_prices['YEAR MONTH'] = pd.to_datetime(district_prices['YEAR MONTH'], errors='coerce')
```

```python
print(district_prices['YEAR MONTH'].dtype)
```

```python
# changing the name of column from YEAR MONTH TO DATE
district_prices.rename(columns= {'YEAR MONTH': 'DATE'}, inplace= True)
```

```python
# this column is not much inconsistant.
# we are replacing these letters bcz of pronunciation of Ё in Russian
district_prices['DISTRICT'] = district_prices['DISTRICT'].str.replace('oye', 'oe', case=False)
district_prices['DISTRICT'] = district_prices['DISTRICT'].str.replace('yovo', 'evo', case=False)
# replacing hyphens in between names to convert them in to standard form
district_prices['DISTRICT'] = district_prices['DISTRICT'].str.replace('-', ' ')
district_prices['DISTRICT'] = district_prices['DISTRICT'].str.capitalize()
```

```python

# Visualization
# 1st of all analyzing data, checking outliers by using boxplot
numeric_columns = district_prices.select_dtypes(include=['int64', 'float64'])

for col in numeric_columns:
  plt.figure(figsize=(6, 3))
  sns.boxplot(data= district_prices, x=col)
  plt.title(f'Checking For Outliers in {col}')
  plt.show()
```

```python
# comparing CBR KEY RATE PCT and AVG MORTGAGE RATE PCT.as both are directly proportional the line should go stright upward
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 4))
sns.scatterplot(data=district_prices, x='CBR KEY RATE PCT', y='AVG MORTGAGE RATE PCT')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
sns.regplot(
    data=district_prices,
    x='CBR KEY RATE PCT',
    y='AVG MORTGAGE RATE PCT',
    scatter_kws={'alpha':0.6},
    line_kws={'color':'red'}
)
plt.title('Relationship Between CBR Key Rate and Avg Mortgage Rate', fontsize=14, pad=15)
plt.xlabel('CBR Key Rate (%)', fontsize=11)
plt.ylabel('Average Mortgage Rate (%)', fontsize=11)
```

```python
sns.set_theme(style="whitegrid")
sns.boxplot(
    data=district_prices,
    x='CBR KEY RATE PCT',
    y='AVG MORTGAGE RATE PCT',
)
plt.title('Relationship Between CBR Key Rate and Avg Mortgage Rate', fontsize=14, pad=15)
plt.xlabel('CBR Key Rate (%)', fontsize=11)
plt.ylabel('Average Mortgage Rate (%)', fontsize=11)
```

```python
plt.figure(figsize=(8, 5))


x_numeric = district_prices['SECONDARY PRICE PER SQM'].astype(float)
y_numeric = district_prices['RENTAL PRICE PER SQM MONTHLY'].astype(float)


sns.scatterplot(x=x_numeric, y=y_numeric, alpha=0.6)

plt.title('Relationship Between Secondary Purchase Price and Rental Price')
plt.xlabel('Secondary Price Per SQM (RUB)')
plt.ylabel('Monthly Rental Price Per SQM (RUB)')

plt.show()
```

```python
target_cols = ['SECONDARY PRICE PER SQM', 'NEWBUILD PRICE PER SQM', 'NEWBUILD MOM CHANGE PCT']

for col in target_cols:
    plt.figure(figsize=(8, 4))
    # Using bins=40 to give a nice, detailed balance to the bars
    sns.histplot(data=district_prices, x=col, bins=40, kde=True, color='crimson')

    plt.title(f'Distribution Shape of {col}')
    plt.show()
```

```python
plt.figure(figsize=(10, 8))

# 1. Calculate the correlation numbers
corr_matrix = district_prices.corr(numeric_only=True)

# 2. Plot it with colors (coolwarm makes negatives blue and positives red)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title('Correlation Matrix of Moscow Real Estate Data')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# 1. Calculate the average price per okrug and sort them descending
okrug_order = district_prices.groupby('OKRUG')['SECONDARY PRICE PER SQM'].mean().sort_values(ascending=False).index

# 2. Plot the bar chart
sns.barplot(
    data=district_prices,
    x='OKRUG',
    y='SECONDARY PRICE PER SQM',
    order=okrug_order,
    palette='Reds_r'
)


plt.title('Average Secondary Market Price per SQM by Okrug (Sorted)')
plt.xlabel('Administrative Okrug')
plt.ylabel('Average Price per SQM (RUB)')
plt.xticks(rotation=45)

plt.show()
```
