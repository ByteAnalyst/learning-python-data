# **Handling missing values**

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/synthetic_dataset.csv')
np.random.seed(0)
df.head()
```

```python
# missing value count
missing_values_count = df.isnull().sum()
print(missing_values_count)
```

```python
# % of missing values
total_cells = np.prod(df.shape)
empty_cells = missing_values_count.sum()

missing_value_percentage = (empty_cells/total_cells) * 100

print(missing_value_percentage)
```

```python
# use df.dropna()if you want to delete missing values.

delete_rows_with_missing_values = df.dropna()
# dlt_category = df.drop(columns=['Category'], inplace=True)
print(delete_rows_with_missing_values)
delete_rows_with_missing_values.head()
```
