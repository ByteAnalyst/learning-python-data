Grouping Data by groupby()

```python
data1 = {
    'name': ['Noor', 'Malaika', 'Laiba', 'Ali', 'Ahmad', 'Laiba', 'Noor'],
    'age': [16, 12, 22, 25, 23, 20, 17],
    'adress': ['Pakistan', 'China', 'Kashmir', 'Gilgit', 'Malacca', 'Izmir', 'Delhi'],
    'education': ['MA', 'BA', 'BSc', 'BFA', 'LLB', 'BEd', 'Bcom']
}

df = pd.DataFrame(data1)
gk = df.groupby('name')
print(gk.first())
```

Aggregate

```python
result = df.groupby('name')['age'].aggregate('sum')
print(result)
```

Scatter Plot

```python
import pandas as pd
import numpy as np

np.random.seed(42)
ages = np.random.randint(18, 80, 50)
augments = (ages // 10) + np.random.randint(0, 5, 50)

df = pd.DataFrame({
    'Subject_Age': ages,
    'Augment_Count': augments,
    'Type': np.random.choice(['Medical', 'Tactical', 'Lifestyle'], 50)
})

df.to_csv('cyber_implants.csv', index=False)

df = pd.read_csv('cyber_implants.csv')
print(df)
```

```python
plt.scatter(df['Type'], df['Augment_Count'])
plt.title('Scatter Plot')
plt.legend('Augment_Count ')
plt.show()
```

Box Plot

```python
plt.boxplot(df['Augment_Count'])
plt.title('Box Plot')
plt.legend('Augment_Count ')
plt.show()
```

Correlation Heatmap

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('cyber_implants.csv')

numeric_df = df.select_dtypes(include=['number'])
# .corr = claculating correlation
corr_df = numeric_df.corr()

plt.imshow(corr_df, cmap='autumn', interpolation='bilinear')
plt.colorbar(label='correlation coefficient')
plt.title('Heat Map')

plt.xticks(range(len(corr_df.columns)), corr_df.columns, rotation=90)
plt.yticks(range(len(corr_df.columns)), corr_df.columns)

plt.tight_layout()
plt.show()
```
