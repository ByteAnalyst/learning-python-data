```python
df = df.drop(columns= 'mentality_composure')
```

```python
# filling missing value
financials = ['value_eur', 'wage_eur', 'release_clause_eur']
df[financials] = df[financials].fillna(0)
```

```python
df[financials].isnull().sum()
```

```python
club = ['club_name', 'league_name', 'club_position']
df[club] = df[club].fillna('Unknown')
```

```python
df[club].isnull().sum()
```

```python
performance_Attributes = ['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']
df[performance_Attributes] = df[performance_Attributes].fillna(df[performance_Attributes].median())
```

```python
df[performance_Attributes].isnull().sum()
```

```python
df['league_level'] = df['league_level'].fillna(1.0)
```

```python
df.isnull().sum()
```

```python
# fixing dtypes
cols_to_fix = ['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic', 'league_level']

df[cols_to_fix] = df[cols_to_fix].astype('int64')
```

```python
df = df.map(lambda x: x.strip() if isinstance(x, str) else x)
```
