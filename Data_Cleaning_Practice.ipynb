```python
import pandas as pd
import numpy as np

df = pd.read_csv('/Data Cleaning1.csv')
df.head()
```

```python
# understaning data
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.replace('\xa0', ' ')
cols = ['Actual gross', 'Adjusted gross (in 2022 dollars)',	'Average gross']
df[cols] = df[cols].apply(lambda x: x.str.replace(r'[\$,]', '', regex=True))
```

```python
# converting some columns having number, from strings to numerics
cols_to_numerics = ['Peak',	'All Time Peak',	'Actual gross',	'Adjusted gross (in 2022 dollars)',	'Average gross',	'Ref.']
df[cols_to_numerics] = df[cols_to_numerics].apply(pd.to_numeric, errors='coerce')
```

```python
df['Start_Year'] = df['Year(s)'].str.split('–|-').str[0]

dates = pd.to_numeric(df['Start_Year'], errors='coerce')
```

```python
# fill the empty places
empty_columns = ['Peak', 'All Time Peak']
df[empty_columns] = df[empty_columns].fillna(df[empty_columns].mean())
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Ref.'])
df['Actual gross'] = df['Actual gross'].fillna(df['Actual gross'].mean())
```

```python
# checking duplicates
df.duplicated().sum()
```

```python
# checking inconsistencies

cols = ['Artist', 'Tour title']
df[cols] = df[cols].apply(lambda x: x.str.strip().str.lower())
```

```python
df['Artist'].unique()
```

```python
df['Tour title'].unique()
```

```python
pattern = r'\[.*?\]|[†‡*]'

df['Tour title'] = df['Tour title'].str.replace(pattern, '', regex=True).str.strip()
```

```python
text_cols = ['Artist', 'Tour title']
df[text_cols] = df[text_cols].apply(lambda x: x.str.title())
```
