Dealing with strings

```python
import pandas as pd
import numpy as np

happiness2015 = pd.read_csv('/content/World_Happiness_2015 (1).csv')
happiness2016 = pd.read_csv('/content/World_Happiness_2016.csv')
happiness2017 = pd.read_csv('/content/World_Happiness_2017.csv')
world_dev = pd.read_csv('/content/World_dev.csv')

```

```python
world_dev = pd.read_csv("World_dev.csv")
col_renaming = {'SourceOfMostRecentIncomeAndExpenditureData': 'IESurvey'}

merged = pd.merge(left=happiness2015, right=world_dev, how='left',
                 left_on='Country', right_on='ShortName')
merged = merged.rename(col_renaming, axis=1)
```

```python
merged[['Country', 'CurrencyUnit']].head()
```

```python
def extract_last_word(element):
  return str(element).split()[-1]
merged['Currency Apply'] = merged['CurrencyUnit'].apply(extract_last_word)
merged['Currency Apply'].head()
```

```python
# using pandas vactorized way
merged['Currency Vectorized'] = merged['CurrencyUnit'].astype(str).str.split().str.get(-1)
merged['Currency Vectorized'].head()
```

```python
uniques = merged['Currency Vectorized'].unique()
uniques.sort()
uniques
```

```python
merged['Currency Vectorized'] = merged['Currency Vectorized'].str.lower()
merged['Currency Vectorized'] = merged['Currency Vectorized'].str.strip()
merged['Currency Vectorized'] = merged['Currency Vectorized'].replace({
    'rubel': 'ruble',
    'rial': 'riyal',
    'sterling': 'pound',
})
```

```python
merged['Currency Vectorized'].head()
```

```python
#working with missing and duplicate data
print(happiness2015.columns.tolist())
```

```python
happiness2015.columns = happiness2015.columns.str.replace('(', '').str.replace(')', '').str.strip().str.upper()
print(happiness2015.columns.tolist())
```

```python
happiness2016.columns = happiness2016.columns.str.replace('(', '').str.replace(')', '').str.strip().str.upper()
happiness2017.columns = happiness2017.columns.str.replace('(', '').str.replace(')', '').str.strip().str.upper()

combined = pd.concat([happiness2015, happiness2016, happiness2017], ignore_index=True)
```

```python
print(combined.notnull().sum().sort_values())
```

```python
print(f'Columns Before: {len(combined.columns)}')
combined = combined.dropna(thresh=159, axis=1)
print(f'Columns After: {len(combined.columns)}')
print(combined.columns.tolist())
```

```python
rows_before = len(combined)
combined = combined.dropna()
print(f"Rows before: {rows_before}")
print(f"Rows after: {len(combined)}")
print(f"Percentage of data lost: {((rows_before - len(combined))/rows_before * 100):.1f}%")
```
