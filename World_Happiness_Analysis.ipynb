```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/World_Happiness_2015.csv')
happiness_index_2015 = pd.read_csv('/content/World_Happiness_2015.csv')
happiness_index_2016 = pd.read_csv('/content/World_Happiness_2016.csv')
happiness_index_2017 = pd.read_csv('/content/World_Happiness_2017.csv')
happiness_index_2015['Year'] = 2015
happiness_index_2016['Year'] = 2016
happiness_index_2017['Year'] = 2017
df.head()
```

```python
# Using aggregation to group data and check some facts.
grouped = df.groupby('Region')
happy_grouped = grouped['Happiness Score']
happy_mean = happy_grouped.mean()
print(happy_mean)
```

```python
# Combining datasets using CONCAT
data_2015 = happiness_index_2015[['Country',	'Happiness Score', 'Year']].head()
data_2016 = happiness_index_2016[['Country',	'Happiness Score', 'Year']].head()

concating_horizontally = pd.concat([data_2015, data_2016], axis= 1)
print(concating_horizontally)

concating_vertically = pd.concat([data_2015, data_2016], axis= 0)
print(concating_vertically)
```

```python
# Combining data sets using MERGE
data_2015_i = happiness_index_2015[['Country',	'Happiness Score', 'Year']].iloc[0:5]
data_2016_ii = happiness_index_2016[['Country',	'Happiness Score', 'Year']].iloc[0:5]

pd.merge(left= data_2015_i, right=data_2016_ii, on= 'Country', suffixes=('_2015', '_2016'))
```

Transformations in pandas

```python
# simple transformation with series.map
def label(element):
  if element > 1:
    return 'High'
  else:
    return 'Low'

happiness_index_2015['Economy Impact'] = happiness_index_2015['Economy (GDP per Capita)'].map(label)
print(happiness_index_2015[['Country', 'Economy (GDP per Capita)', 'Economy Impact']].iloc[5:11])
```

```python
# Using series.apply for flexablility
def label(element, x):
  if element > x:
    return 'High'
  else:
    return 'Low'
happiness_index_2015['Economy Impact'] = happiness_index_2015['Economy (GDP per Capita)'].apply(label, x=0.8)
```

```python
# analyzing how each column participate in happiness
factors = ['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual']
def percentages(col):
  div = col/happiness_index_2015['Happiness Score']
  return div*100
factors_percentage = happiness_index_2015[factors].apply(percentages)
factors_percentage.head()
```
